```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(1)
```

```python
df = pd.read_csv('oscars.csv')
pd.set_option('display.max_columns', 1000)
pd.set_option('display.max_rows', 1000)
```

```python
df.head(1000)
```

```python
years = df.loc[((df['category']=='BEST PICTURE') | 
        (df['category']=='OUTSTANDING PICTURE')) & 
       (df['winner'] == True), ['entity', 'year']]

best_pics = df.loc[((df['category']=='BEST PICTURE') | 
        (df['category']=='OUTSTANDING PICTURE')) & 
       (df['winner'] == True), ['entity']]

actors = df.loc[((df['category']=='ACTOR IN A LEADING ROLE') | 
        (df['category']=='ACTRESS IN A LEADING ROLE')) & 
       (df['winner'] == True), ['entity', 'year']]
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
years = years[2:]
best_pics = best_pics[2:]
```

```python
best_pics
```

```python
best_pic_title_list = []
for row in best_pics['entity']:
    best_pic_title_list.append(len(row))
    
actor_len_list = []
for row in actors['entity']:
    actor_len_list.append(len(row))
    
actor_len_list = actor_len_list
# [28:]
len(actor_len_list)
```

```python
plt.scatter(years, best_pic_title_list)
plt.title('Relationship Between Years and Length of Best Picture Title \n Correlation = .082')
plt.xlabel('Year')
plt.ylabel('Length of Best Picture Title')
plt.show()
```

```python
double = []
for year in years['year']:
    double.append(year)
double.extend(double)
double_years = sorted(double)[28:]
len(double_years)
```

```python
plt.scatter(double_years, actor_len_list)
plt.title('Relationship Between Years and Length of Best Actor/Actress Names \n Correlation = .204')
plt.xlabel('Year')
plt.ylabel('Length of Best Actor/Actress Names')
plt.show()
```

```python
np.corrcoef(actor_len_list, double_years)
```
